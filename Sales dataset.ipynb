```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Customer_ID': [201, 202, 203, 204, 205, 206],
    'Cname': ['Rohit', 'Mohit', 'Sita', 'Neha', 'Ram', 'Priya'],
    'PName':['Facewash','Biscuit','Toothpaste','Notebook',np.nan,'Soap'],
    'Quantity': [2, 6, 1,9, np.nan, 3],
    'Cost': [150, 250, 200, 600, 700, 90],
    'Mode_of_payment': ['Online', 'Cash', np.nan, 'Card', 'Online', 'Online']
}
df=pd.DataFrame(data)
df
```

```python
df['PName']=df['PName'].fillna('Chocolate')
df['Quantity']=df['Quantity'].fillna(7)
df['Mode_of_payment']=df['Mode_of_payment'].fillna('Cash')
df
```

```python
result=df[df['Cost']>400]
print(result)
```

```python
plt.plot(df['Customer_ID'], df['Cost'],marker='o',linestyle='--',color='green')

plt.xlabel('Customer ID')
plt.ylabel('Cost')
plt.title('Cost vs Customer ID')
plt.grid(True)
plt.show()
```

```python
plt.hist(df['Cost'], bins=5, color='purple')

plt.xlabel('Cost')
plt.ylabel('Frequency')
plt.title('Histogram of Cost')
plt.show()
```

```python
plt.scatter(df['Quantity'], df['Cost'],
            marker='x',
            color='red')

plt.xlabel('Quantity')
plt.ylabel('Cost')
plt.title('Quantity vs Cost')
plt.show()
```
