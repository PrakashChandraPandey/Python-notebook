```python
import pandas as pd
import numpy as np

data = {
    'PID': [101, 102, 103, 104, 105],
    'Age': [25, np.nan, 45, 35, np.nan],
    'Gender': ['M', 'F', 'M', 'F', 'M'],
    'Disease': ['Diabetes', 'BP', 'Heart', 'BP', 'Diabetes'],
    'Income': [30000, 45000, np.nan, 50000, 40000],
    'Weight': [70, 60, 80, np.nan, 75],
    'BP': [120, 130, np.nan, 60, 180]
}

df = pd.DataFrame(data)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Income'] = df['Income'].fillna(df['Income'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
df['BP'] = df['BP'].fillna(df['BP'].mean())

print(df)
```

```python
print(df[df['Income'] < 40000])
```

```python
plt.plot(df['Age'], df['BP'],
         marker='o')

plt.xlabel("Age")
plt.ylabel("BP")
plt.title("Age vs BP")
plt.show()
```

```python
plt.scatter(df['Weight'], df['BP'],
            color='blue',
            marker='o')

plt.xlabel("Weight")
plt.ylabel("BP")
plt.title("Weight vs BP")
plt.show()
```

```python
plt.bar(df['PID'], df['Income'],
        color='green')

plt.xlabel("PID")
plt.ylabel("Income")
plt.title("Income Bar Chart")
plt.show()
```

```python
plt.hist(df['BP'],
         bins=5,
         color='purple')

plt.xlabel("BP")
plt.ylabel("Frequency")
plt.title("BP Histogram")
plt.show()
```

```python
plt.pie(df['Age'],
        labels=df['Disease'],
        autopct='%1.1f%%')

plt.title("Disease Wise Age Distribution")
plt.show()
```

```python
sns.lineplot(x='Age', y='BP',
             data=df,
             linestyle='dashed',
             marker='o',
             color='orange')

plt.show()
```

```python
sns.scatterplot(x='Income',
                y='BP',
                data=df,
                color='black')

plt.show()
```

```python
sns.barplot(x='Gender',
            y='Income',
            data=df)

plt.show()
```

```python
corr = df[['Age','Income','Weight','BP']].corr()

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.show()
```

```python
sns.boxplot(y='BP',
            data=df,
            color='cyan')

plt.title("BP Outlier Detection")
plt.show()
```
