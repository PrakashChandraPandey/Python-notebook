```python
df["Car_Model"]=df["Car_Model"].fillna("Alto")
df["Type_of_Car"]=df["Type_of_Car"].fillna("CNG")
df["Price"]=df["Price"].fillna(df["Price"].mean())
df["Color"]=df["Color"].fillna("Black")
df["Mileage"]=df["Mileage"].fillna(df["Mileage"].mean())
df["Engine_CC"]=df["Engine_CC"].fillna(df["Engine_CC"].mean())
df["Manufacturer"]=df["Manufacturer"].fillna("Mahindra")
```
